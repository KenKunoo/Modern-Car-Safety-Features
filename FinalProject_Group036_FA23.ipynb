# Research Question

## Background and Prior Work

With each passing year, more safety technologies are developed and implemented into newer generations of cars. Safety features such as backup cameras and automatic braking systems have steadily become a common sight over the course of the past few years and with the implementation of these changes, we also see a change in driver behavior. Previous works have examined the effects and effectiveness of these safety systems, such as:

Advanced driver assistance: These systems encompass a variety of functionalities, such as front crash prevention, lane departure warnings, blind-spot detection, and rear crash prevention. Detailed information on how these technologies contribute to reducing crashes and improving safety can be found on the Insurance Institute for Highway Safety (IIHS) website.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Pedestrian Detection Systems: In a study focused on pedestrian safety, Cicchino has highlighted the effectiveness of automatic braking systems that detect pedestrians. According to this research, automatic braking systems that recognize pedestrians cut pedestrian crashes by 27 percent <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2).

There also exists another category of studies that delve into the effects of vehicle characteristics on the behaviors of drivers. 3 studies, making use of naturalistic observation techniques as well as questionnaires, conducted with UK drivers found that the higher the perceived performance of a car is, defined as having a higher horse power, breaking power, etc., the more likely drivers are to engage in more risky driving, defined as drivers driving at a compariatively higher speed than if they were driving in a car with a lower perceived performance as well as committing more driving violations in a car of higher performance than a car of lower performance.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) It is to be noted that this study did not include modern makes and models of cars and only included models of cars from the 1990s. However, this study is still relevant background to be aware of since it establishes a predence for the notion that drivers are more likely to behave in risky fashion when driving a vehicle with a higher level of perceived performance. 

This study on UK drivers had also examined the effects of the amount of safety features on driver behavior. The study identified 4 characteristics of cars, those being smoothness of handling, performance, amount of safety features, and internal noise levels, and formulated questionnaires describing hypothetical cars with varying levels of each characteristic. Drivers responses on intended driving speed and driving violations were collected using the same methods as the previous set of questionnaires above. The results seemed to indicate the more safety features a car is described to have, the faster drivers would intend to drive.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

## Crash Reporting - Drivers Data

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
crash_report = pd.read_csv('crash_report.csv', low_memory=False)
```

```python
crash_report.head()
```

```python
list(crash_report.columns)
```

```python
# Removing unnecessary columns like report number, local case number and etx.
crash_report = crash_report.drop(columns = ['Report Number', 
                   'Local Case Number',
                   'Agency Name',
                    'Route Type',
                   'Road Name',
                   'Cross-Street Type',
                   'Cross-Street Name',
                   'Off-Road Description',
                   'Municipality',
                   'Related Non-Motorist',
                   'Person ID',
                   'Drivers License State',
                   'Vehicle ID',
                   'Vehicle Damage Extent',
                   'Vehicle First Impact Location',
                   'Vehicle Second Impact Location',
                    'Vehicle Continuing Dir',
                    'Vehicle Going Dir',
                   'Latitude', 
                   'Longitude',
                   'Location'])
crash_report.head()
```

```python
# Checking NaN values in Vehicle Make
crash_report[crash_report['Vehicle Make'].isna() == True]

# note: 24 NaN values
# 168515 - 24 = 168491
```

```python
# Removing all the NaN values in Vehicle Make
crash_report = crash_report.dropna(subset=['Vehicle Make'])
crash_report.shape
```

```python
# Looking at the values of vehicle year some values are too small such as 10 and some values are too large such as 9999
# So we will trim down the data to only fit the years 1980 to 2023

crash_report = crash_report[(crash_report['Vehicle Year'] >= 1980) & (crash_report['Vehicle Year'] <= 2023)]
crash_report.shape
```

```python
# Vehicle Body type: we only want to look at cars, since in the data there are Emergency Vehicles and Semi trucks
crash_report = crash_report[(crash_report['Vehicle Body Type'] == '(SPORT) UTILITY VEHICLE') | 
             (crash_report['Vehicle Body Type'] == 'PASSENGER CAR') | 
             (crash_report['Vehicle Body Type'] == 'RECREATIONAL VEHICLE') |
             (crash_report['Vehicle Body Type'] == 'PICKUP TRUCK') |
             (crash_report['Vehicle Body Type'] == 'STATION WAGON') |
             (crash_report['Vehicle Body Type'] == 'ALL TERRAIN VEHICLE (ATV)') |
             (crash_report['Vehicle Body Type'] == 'VAN')]
crash_report.shape
```

Some of the Vehicle make have spelling errors, so we will be cleaning the spelling error
For example Toyota was written as 'Toyota', 'Toyt'

Data Cleaning:
Our data is cleaned to improve quality of life, but we still kept a few potential columns that we may or may not need. In order to clean the data, we first omitted any unnecessary columns, such as case number, location name, location position, driver identification, and vehicle damage details. Then, we removed any car manufacture that is NaN, so we can identify the vehicle. The car manufacture also has typos, so we fixed the typos and shorten the name to first four letters to create uniformity. ### Exploring Driver Fault Between Older and Newer Model Cars

```python
#Let's check out the unique values in the Vehicle Year column
#This will make it easier for us to decide where to split older and newer cars

crash_report['Vehicle Year'].unique()
```

It looks like we have a good amount of vehicle years from the 80s/90s as well as the 2010s and 2020s. We'll split the data at 2009, defining any cars that were made in and before 2009 as old cars while our new car set will consist of cars that were made in 2010 going forward. 2009 also serves as the split year since widespread use of automatic emergency breaking systems began in 2010 (https://www.robsonforensic.com/articles/automatic-emergency-braking-expert). 

```python
older_models = crash_report[crash_report['Vehicle Year'] < 2010]
newer_models = crash_report[crash_report['Vehicle Year'] >= 2010]
```

Right away, we can see that our data set contains much more newer models of cars than older models, seeming to run contrary to our hypothesis. This begs the question: is this because of the implementation of these safety features? We'll start answering this question by taking a look at and comparing the proportion of drivers at fault between the older and newer models of cars. We will be using the proportions to compare the data sets due to this massive difference in the sizes of the data sets.

```python
# Fault Proportions for older models of cars
older_fault = older_models['Driver At Fault'].value_counts(normalize=True).reset_index()
older_fault
```

```python
# Fault Proportions for newer models of cars
newer_fault = newer_models['Driver At Fault'].value_counts(normalize=True).reset_index()
newer_fault
```

```python
# Propotion values for older model cars whose drivers were at fault, not at fault, and where fault is unknown, respectively
old_yes = older_fault.iloc[0,1]
old_no = older_fault.iloc[1,1]
old_un = older_fault.iloc[2,1]

# Proportion values for newer model cars whose drivers were at fault, not at fault, and where fault is unknown, respectively
new_yes = newer_fault.iloc[0,1]
new_no = newer_fault.iloc[1,1]
new_un = newer_fault.iloc[2,1]

# Combine fault proportions between older and newer models of cars into one dataframe
fault = pd.DataFrame()

fault['Model'] = ['Older','Older','Older','New','New','New']
fault['Fault'] = ['Yes', 'No', 'Unknown', 'Yes', 'No', 'Unknown']
fault['Proportion'] = [old_yes, old_no, old_un, new_yes, new_no, new_un]

fault
```

```python
sns.barplot(data=fault, x='Fault', y='Proportion', hue='Model')
```

After compiling and organizing the fault proportion data, we can now see from the bar plot above that a higher proportion of drivers of older model cars were at fault in crashes than their counterparts who drive newer model cars. This pattern seems to suggest that the presence of advanced safety systems, such as automatic emergency braking systems, seems to reduce the rate at which drivers cause accidents themselves.

### Exploring Driver Substance Abuse Patterns between Drivers of Older and Newer Models of Cars

Exploring the relationship between driver substance abuse patterns and the age of their vehicles presents an intriguing facet of our road safety project. We are particularly interested in understanding whether the advanced safety features in newer car models influence the prevalence of substance abuse among drivers. The rationale behind this inquiry stems from the known risks associated with substance-impaired driving, which often leads to dangerous behavior on the roads. Therefore, investigating whether these modern safety systems have any impact on the frequency of substance abuse among drivers of newer versus older cars is not only relevant but also crucial for enhancing our understanding of road safety dynamics. This exploration could reveal significant insights into how vehicle technology interacts with driver behavior, especially in the context of substance abuse, thereby contributing to a more comprehensive approach to road safety.

```python
crash_report['Driver Substance Abuse'].unique()
```

We are able to know the substance abuse factor.

```python
# Proportions of substance abuse for older and newer models
old_substance = older_models['Driver Substance Abuse'].value_counts(normalize=True).reset_index()
old_substance.columns = ['Substance', 'Proportion']
old_substance
```

```python
new_substance = newer_models['Driver Substance Abuse'].value_counts(normalize=True).reset_index()
new_substance.columns = ['Substance', 'Proportion']
new_substance
```

Here's the proportion of each substance abuse.

```python
old_substance_proportion = old_substance['Proportion']
new_substance_proportion = new_substance['Proportion']

# Combine
combined_substance = pd.concat([old_substance, new_substance], keys=['Older', 'New'])
combined_substance.reset_index(inplace=True)
combined_substance.drop('level_1', axis=1, inplace=True)
combined_substance.rename(columns={'level_0': 'Model'}, inplace=True)
combined_substance
```

We noticed that the main protion is NONE DETECTED and UNKNOWN. So we combined the rest which is DETECTED together, in order to see the portion of accident that was affected by substance abuse.

```python
def categorize_substance(row):
    if row['Substance'] in ['NONE DETECTED', 'UNKNOWN']:
        return row['Substance']
    else:
        return 'Detected'

combined_substance['Category'] = combined_substance.apply(categorize_substance, axis=1)
grouped_data = combined_substance.groupby(['Model', 'Category'])['Proportion'].sum().reset_index()

sns.barplot(data=grouped_data, x='Category', y='Proportion', hue='Model')
plt.title('Proportion of Substance Abuse Categories in Older and Newer Cars')
plt.show()
```

The barplot above shows that the proportion for the driver to be detected as substance abuse is very small, it's lower than 0.1. And there is barely no differences between newer and older model.

### T-Test for T-Test for 'Vehicle Year' by 'Driver At Fault'

```python
sns.set(style="whitegrid")
crash_report = crash_report[(crash_report['Vehicle Year'] >= 1980) & (crash_report['Vehicle Year'] <= 2023)]
# T-Test: Mean 'Vehicle Year' for 'Driver At Fault' vs. 'No Driver At Fault'
grouped_fault = crash_report.groupby('Driver At Fault')['Vehicle Year']
t_stat, p_value = ttest_ind(grouped_fault.get_group('Yes'), grouped_fault.get_group('No'))
print(f"Vehicle Year - Driver At Fault vs No Fault: t-statistic={t_stat}, p-value={p_value}")

# Visualization for 'Vehicle Year' by 'Driver At Fault'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Driver At Fault', y='Vehicle Year', data=crash_report, hue='Driver At Fault', palette=['pink', 'lightblue', 'lightgreen'], dodge=False)
plt.title('Vehicle Year by Driver At Fault Status')
plt.xlabel('Driver At Fault')
plt.ylabel('Vehicle Year')
plt.show()
```

The T-Statistic is approximately -32.400, and the P-Value is virtually 0 (1.51e-229). Since the p-value is much less than 0.05, this indicates that we have extremely strong evidence that there is a statistically significant difference in the mean 'Vehicle Year' between the groups where the driver was at fault and where the driver was not at fault.
